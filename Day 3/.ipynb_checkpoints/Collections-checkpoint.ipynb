```python
# Collections 
# Lists

# Ordered, Mutable, Allow Duplicate Values

myList = [10,20,30,40,50]

print(myList)
print(myList[3]) # access based on index

myList.append('Paras') # append add element at last index
print(myList)

myList.insert(2,25)
print(myList)

print(myList.pop()) # return the last element and remove the last value
print(myList)

print(f"The Length of the list is : {len(myList)}")

print("Maximum Value :", max(myList))
print("Minimum Value :", min(myList))
print("Total Sum :", sum(myList))

# Remove the value by index
myList.remove(25)
print(myList)

# Check membership
print(20 in myList)

mlist =[10,20,30,40,50]
print(myList is mlist)

print(myList[1:4])

myList.extend([1,2,3,4,5]) # adding element at last
print(myList)

# Delete using index number
del myList[1]
print(myList)

myList.clear()
print(myList)
```

```python
# Loops in LISTs

myData = [1,2,3,4,5,6,7,8]

for num in myData:
    print(num, end = " ")
print()    

i = 0
while i < len(myData):
    print(myData[i] , end = " ")
    i += 1
print()    

myData.reverse()
print(myData)

myData.sort()
print(myData)

myData.sort(reverse = True)
print(myData)

# Copy the List

copy_number = myData.copy()
print(copy_number)

# Sort Way to Create List

squares = [x ** 2 for x in range(1,6)]
print(squares)
```

```python
# Tuple

# Ordered, Immutable

fruits = ("Orange", "Apple", "Banana")
print(fruits)

mixed = ('Paras', 21, 'Ambala', 67.33, True)
print(mixed)

# Empty Tuple
empty_tup = ()
print()

# Tuple with single value
data = ('hell',)
print(data)
print(type(data))

# Access value in tuple by index
print(fruits[2])
```

```python
# As Tuple is immutable, then for modifying the tuple 
# We have to convert tuple into list 
# Then Modify the List
# Lastly, Convert the list into tuple

number_tuple = (23, 'Paras', 78.33)
print(number_tuple)

number_tuple = list(number_tuple)
print(number_tuple)

number_tuple.append('Harikesh')
print(number_tuple)

number_tuple = tuple(number_tuple)
print(number_tuple)
```

```python
# Set -- Unordered , Unique Elements, Mutable, Supports Set Operations(Union, Intersection)

names = { "alex", "bob", "john", "catty"}
print(names)

for val in names:
    print(val)

names.add('krishna')
print(names)

# To add many elements
names.update(['rohit', 'pawan'])
print(names)

# Removing the value from set
names.remove('rohit')
print(names)

# names.remove('parad')
names.discard('parad')

names.pop();
print(names)
```

```python
# Set Operations

A = {1,2,3,4,5}
B = {4,5,6,7,8}

# Union
print(A|B)
print(A.union(B))

# Intersection
print(A & B)
print(A.intersection(B))

# Difference
print(A - B)
print(A.difference(B))

# Symmetric Difference
print(A ^ B)
print(A.symmetric_difference(B))
```

```python
# Task 1

def add_employee(name, salary, email):
    data.append({"name" : name,"salary" : salary,"email" : email  })
    print(data)

def printEmployees():
    print("Details : ------------------------------------------------------------------------------------------------------------------------------")
    for items in data:
        for key,value in items.items():
            print(key , ":", value)
        print("----------------------------------------------------------------------------------------------------")

def updateEmployees(name, updatedname, updatedsalary, updatedemail):
    for emp in data:
        if emp['name'] == name:
            emp['name'] = updatedname
            emp['salary'] = updatedsalary
            emp['email'] = updatedemail
            print("Details Updated Successfully") 
            return

def delete_employee(name):
    for emp in data:
        if emp['name'] == name:
            data.remove(emp)
            print(f"Deleted Data :\nName : {emp['name']}\nSalary : {emp['salary']}\nEmail : {emp['email']}")
            return

data = [
    {
    "name" : "Paras",
    "salary" : 22000,
    "email" : "paras@gmail.com"  
    },
    {
    "name" : "Harikesh",
    "salary" : 25000,
    "email" : "harikesh@gmail.com"  
    },
    {
    "name" : "Adity",
    "salary" : 27000,
    "email" : "adity@gmail.com"  
    }    
]

print(data)
print("----------------------------------------------------------------------------------------------------")
add_employee("Pragya", 33000, "pakhi@gmail.com")

print("----------------------------------------------------------------------------------------------------")
updateEmployees("Paras", "Paras", 35000, "paras@gmail.com")
print(data)
print("----------------------------------------------------------------------------------------------------")
delete_employee("Harikesh")
print(data)
printEmployees()
```

```python
# Task 2 

movies = []
theaters = set()
ticket_price = (250,350,300)
customers = {}

def add_movie(title, genere, showtime):
    movies.append({'title' : title, 'genere' : genere, 'showtime' : showtime})

    print(f"Added Movie -------------------------------------------\nTitle : {title}\nGenere : {genere}\nShowtime : {showtime}\n")
    return

def remove_movie(title):
    for movie in movies:
        if(movie['title'] == title):
            movies.remove(movie)
            print(f"Removed Movie --------------------------------------\nTitle : {title}\nGenere : {movie['genere']}\nShowtime : {movie['showtime']}\n")
    return        

def add_theater(theater_name):
    theaters.add(theater_name)

    print("Theater Added :",theater_name)

def add_customer(customer_name):
    customers[customer_name] = []
    return

def book_ticket(customer_name, movie_title):
    for movie in movies:
        if movie["title"] == movie_title:
            customers[customer_name].append(movie_title)
            print(f"Ticket booked for '{customer_name}' to watch '{movie_title}'.")
            return

def print_data():
    print("\nMovies Available:")
    for movie in movies:
        print(f"  {movie['title']} ({movie['genere']}) at {movie['showtime']}")

    print("\nTheaters Available:")
    for theater in theaters:
        print(f"  {theater}")

    print("\nTicket Pricing:")
    print(f"  Prices: {ticket_price}")

    print("\nCustomer Bookings:")
    for customer, bookings in customers.items():
        print(f"  {customer}: {', '.join(bookings) if bookings else 'No bookings yet'}")

add_movie("Krrish", "Action", "21:20")
add_movie("Juman Ji", "Adventurous", "15:10")
add_movie("Total Recall", "Sci-Fi", "10:50")

# remove_movie("Juman Ji")

add_theater("PVR")
add_theater("Galaxy")
add_theater("Minerva")

print("---------------------------------------------------------------------------------------------")

add_customer("Adity")
add_customer("Paras")
add_customer("Harikesh")


book_ticket("Adity", "Krrish")
book_ticket("Paras", "Total Recall")
book_ticket("Harikesh", "Juman Ji")

print("---------------------------------------------------------------------------------------------")

print_data()
```
