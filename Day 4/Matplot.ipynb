```python
import matplotlib.pyplot as plt
print(plt.__package__)
```

```python
plt.figure(figsize=(5,6))
```

```python
plt.plot([1,2,3],[4,5,6])
plt.show()
```

```python
x = [1,2,3,4,5]
y = [10,20,25,30,40]

plt.plot(x,y,marker = 'o', linestyle = '--' , color = 'b')
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("Simple Line Plot")
plt.grid(True)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [10,20,25,30,40]

plt.scatter(x,y,marker = '*', color = 'r')
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("Scatter Line Plot")
plt.show()
```

```python
categories = ['A','B', 'C', 'D']
values = [10,20,25,30]

plt.bar(categories , values, color = 'cyan')
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart")
plt.show()
```

```python
# Histogram
import numpy as np
data = np.random.randn(1000)
plt.hist(data,bins = 30, color = 'purple', edgecolor = 'black')
plt.xlabel ("Bins")
plt.ylabel("Frequency")
plt.title("Histogram Example")
plt.show()
```

```python
# Pie Chart

sizes = [20,30,25,25]
labels = ["Adity", "Paras", "Pragya", "Harikesh"]
colors = ["red", "yellow", "cyan", "green"]
explode = (0,0.1,0,0)
plt.pie(sizes, labels = labels, colors=colors, autopct = '%1.1f%%', explode=explode,shadow=True, startangle=140)
plt.show()
```
