```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.DataFrame(
    {
        "x" : [1,2,3,4,5],
        "y" : [10,20,15,25,30]
    }
)

sns.lineplot(x = "x", y = "y", data = data)
plt.title("SeaBorn line Plot")
plt.show()
```
